```
# 2. Path Sum
# Reference: https://leetcode.com/problems/path-sum/
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def hasPathSum1(node, s): 
    # Return true if we run out of tree and s = 0  
    if node is None: 
        return False 
  
    else: 
        ans = 0 
          
        # Otherwise check both subtrees 
        subSum = s - node.val 
          
        # If we reach a leaf node and sum becomes 0, then  
        # return True  
        if(subSum == 0 and node.left == None and node.right == None): 
            return True 
  
        if node.left is not None: 
            ans = ans or hasPathSum1(node.left, subSum) 
        if node.right is not None: 
            ans = ans or hasPathSum1(node.right, subSum) 
  
        return ans 
s= 20
root=TreeNode(5)
root.left = TreeNode(4)
root.right =TreeNode(8)
root.left.left = TreeNode(11)
root.right.left = TreeNode(13)
root.right.right = TreeNode(4)
print(hasPathSum1(root,s))
```
